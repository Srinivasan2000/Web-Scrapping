```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page = requests.get("https://www.vastrado.com/vastrado-men")
soup = BeautifulSoup(page.content, 'html.parser')
page
soup
```

```python
shirts = soup.find(class_="product-name").get_text()
shirts
```

```python
shirts_name = soup.select(".product-name")
shirts_name
```

```python
product_name = soup.select('.brandproduct-title')
product_name
```

```python
prod_name=[]
for i in product_name:
    prod_name.append(i.get_text())
print(prod_name)
```

```python
len(prod_name)
```

```python
shirts=[]
for i in shirts_name:
    shirts.append(i.get_text(strip=True))

print(shirts)
```

```python
len(shirts)
```

```python
rate = soup.select(".price-new")
rate
```

```python
price=[]
for i in rate:
    price.append(i.get_text())

print(price)
```

```python
len(price)
```

```python
old_rate = soup.select('.price-old')
old_rate
```

```python
old_price=[]
for i in old_rate:
    old_price.append(i.get_text())

print(old_price)
```

```python
len(old_price)
```

```python
offer = soup.select('.sale.sale-percentage')
offer
```

```python
off = []
for i in offer:
    off.append(i.get_text())
print(off)
```

```python
len(off)
```

```python
import pandas as pd
vastardo = pd.DataFrame({
    "Brand_name" : prod_name,
    "Product_description" : shirts,
    "Discount_Price" : price,
    "Original_Price": old_price,
    "offers": off
})
vastardo
```

```python
vastardo.to_csv('vastardo.csv',index=False)
```
